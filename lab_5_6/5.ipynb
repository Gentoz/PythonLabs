```python
# 2.7) фильтрация набора данных по какому-либо условию
# Отсортируйте игроков по убыванию роста и найдите пять человек, с самым большим ростом
df1 = pd.read_csv('nba.csv')
df1.sort_values(by='Age', inplace=True, ascending=False)
df1
```

```python
# 2.8) Удаление строк с пропущенными значениями
df2 = pd.read_csv('nba.csv')
df2['Salary'] = df2['Salary'].astype(str)
df2 = df2.loc[(df2['Salary'] != "NaN") & (df2['Salary'] != "nan")]
df2
```

```python
# 2.9) создание нового поля вычисленного на основе значений других полей:
# 2.9.1) через выражение на базе имеющихся колонок
df1 = pd.read_csv('nba.csv')
df1['Age'] = df1['Age'].astype(str)
age_list = list(df1['Age'])
age_sep = []
for i in age_list:
    if float(i) > 25:
        age_sep.append("Old")
    else:
        age_sep.append("Young")
df1['Old?'] = age_sep
df1
```

```python
# 2.9.2) через DataFrame.apply
df1 = pd.read_csv('nba.csv')
df1['Position'] = df1['Position'].astype(str)
def function(row):
    if 'PG' in row['Position']:
        result = (row['Salary'] * 5)
        return result
df1['Salary*5'] = df1.apply(function, axis=1)
df1
```

```python
# 2.9.3) через Series.apply
df1 = pd.read_csv('nba.csv')
df1['Position'] = df1['Position'].astype(str)
gender_list = list(df1['Position'])
a = pd.Series(gender_list)
def func(value):
    if value == "PG":
        return True
    return False
a.apply(func)
```

```python
# 2.10) сортировка по какому-либо из полей
# 2.11) вычислить несколько статистик по колонкам.
df1 = pd.read_csv('nba.csv')
mean_salary = df1['Salary'].mean()
min_salary = df1['Salary'].min()
max_salary = df1['Salary'].max()
print("Среднее значение: ", mean_salary)
print("Минимальное значение: ", min_salary)
print("Максимальное значение: ", max_salary)
```

```python
# 2.14) Удалите текущий индекс и создайте новый индекс на базе новой колонки, которая для этого лучше всего подходит
df1 = pd.read_csv('nba.csv')
df1['Position'] = df1['Position'].astype(str)
def function(row):
    if 'PG' in row['Position']:
        result = (row['Salary'] * 5)
        return result
df1['Salary*5'] = df1.apply(function, axis=1)
df1.set_index('Salary*5', inplace= True )
df1
```
